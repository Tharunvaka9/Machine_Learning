```python
sns.distplot(df['math score'])
plt.show()
#Distplots are primarily used for 
#exploratory data analysis to gain insights 
#into the characteristics of a single variable, such as whether it's normally distributed, skewed, or has multiple modes. 
```

```python
sns.jointplot(x='writing score',y='reading score',data=df,kind='scatter')
plt.show()
#Jointplots are useful for understanding the relationship between two variables,
#visualizing their individual distributions,
#and potentially identifying correlations or patterns. 
```

```python
sns.pairplot(df)
plt.show()
#A pair plot, also known as a scatterplot matrix, is a matrix of graphs that enables the
#visualization of the relationship between each pair of variables in a dataset.
#It combines both histogram and scatter plots,
#providing a unique overview of the dataset's distributions and correlations.
```

```python
sns.pairplot(df,hue='gender')
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
#Heatmaps are useful for exploring data and understanding relationships between variables.
```
